## Preprocessing

```python
Y_df = pd.DataFrame(Y)
Y_df['index'] = range(len(Y_df))
Y_df.rename(columns={0:"Y_df"},inplace=True)


Y_s_df = pd.DataFrame(Y_s)
Y_s_df['index'] = range(len(Y_s_df))
Y_s_df.rename(columns={0:"Y_s_df"},inplace=True)

joind = Y_df.merge(Y_s_df,on='index')
```

```python
joind.head()
```

```python
from sklearn.preprocessing import MinMaxScaler

tr_temp = joind[joind['Y_s_df'] >=5]
Tr_temp_X = X[list(tr_temp.index)]
Tr_temp_X = Tr_temp_X.flatten()
```

```python
scaler = MinMaxScaler(feature_range=(-1,1))
scaler = scaler.fit(Tr_temp_X.reshape(-1,1))
```

### Normalizing

```python
X = scaler.transform(X.flatten().reshape(-1,1)).reshape(X.shape)
```

## Spliting Data

```python
training = joind[joind['Y_s_df'] >=5]
Val = joind[joind['Y_s_df'].isin([6,7])]
test = joind[joind['Y_s_df'].isin([8,9,10])]

Training_X = X[list(training.index)]
Val_X = X[list(Val.index)]
Test_X = X[list(test.index)]


Training_Y = np.asarray(training.Y_df)
Val_Y = np.asarray(Val.Y_df)
Test_Y = np.asarray(test.Y_df)
```

## Training

```python
from data import MRDataModule, MRData
from model import LSTMClassifier, FeatureExtractor
import torch.nn as nn
import torch
from torch.utils.data import DataLoader
from pytorch_lightning.loggers import TensorBoardLogger
from pytorch_lightning import Trainer
from pytorch_lightning.callbacks import ModelCheckpoint
```

### 1 layer, 2000 epochs, non bidirectional

```python
p = dict(
    batch_size = 256, 
    criterion = nn.CrossEntropyLoss(),
    max_epochs = 2000,
    n_features = 60,
    hidden_size = 24,
    num_layers = 1,
    dropout = 0.5,
    learning_rate = 0.001,
    bi_directional = False
)

tb_logger = TensorBoardLogger("tb_logs", name="lstm-1layer-nonbidirectional")

trainer = Trainer(
    accumulate_grad_batches=p['batch_size'],
    max_epochs=p['max_epochs'],
    logger=tb_logger,
    log_every_n_steps=1,
    devices=1,
    num_nodes=-1,
    # gpus=0,
)

model = LSTMClassifier(
    n_features = p['n_features'],
    hidden_size = p['hidden_size'],
    batch_size = p['batch_size'],
    criterion = p['criterion'],
    num_layers = p['num_layers'],
    dropout = p['dropout'],
    learning_rate = p['learning_rate'],
    bidirectional = p['bi_directional']
)

dm = MRDataModule(
    Training_X, Training_Y, Val_X,Val_Y, Test_X, Test_Y,
    batch_size = p['batch_size'],
    num_workers=4,
)
```

```python
trainer.fit(model, datamodule=dm)
```

```python
trainer.save_checkpoint("models/lstm-1layer-nonbidirectional.ckpt")
```

```python
trainer.test(ckpt_path="models/lstm-1layer-nonbidirectional.ckpt",datamodule=dm)
```

### Pca Analysis

### Taking lstm 1 layer Non bidirectional since it's validation loss is comparatively low 

```python
p = dict(
    batch_size = 256, 
    criterion = nn.CrossEntropyLoss(),
    max_epochs = 2000,
    n_features = 60,
    hidden_size = 24,
    num_layers = 1,
    dropout = 0.5,
    learning_rate = 0.001,
    bi_directional = False
)
```

```python
model_loded = LSTMClassifier.load_from_checkpoint("models/lstm-1layer-nonbidirectional.ckpt",n_features = p['n_features'],
    hidden_size = p['hidden_size'],
    batch_size = p['batch_size'],
    criterion = p['criterion'],
    num_layers = p['num_layers'],
    dropout = p['dropout'],
    learning_rate = p['learning_rate'],
    bidirectional = p['bi_directional'],)

features = FeatureExtractor(model_loded, layers=['lstm'])
```

```python
features_all = features(torch.from_numpy(Training_X.astype('float32')))
lstm = features_all['lstm'][1][0][0]


from sklearn.decomposition import PCA

pca = PCA(n_components=2)
reduced = pca.fit_transform(lstm.detach().numpy())
```

```python
dict_data = {}
for data, class_lb in zip(reduced, Training_Y):
        if dict_data.get(class_lb) == None:
            dict_data[class_lb] = []
        dict_data[class_lb].append(list(data))
```

```python
plt.figure().set_figwidth(15)
plt.figure().set_figheight(5)
for key, value in dict_data.items(): 
    value = np.asarray(value)
    plt.scatter(value[:,0], value[:,1], label=key)
plt.legend(dict_data.keys(),loc='center right',bbox_to_anchor=(1.18, 0.5))    
```
